# Boxplot
A boxxplot is a standardized way of displaying the distrubution od data based on a five number summry ("minimum",first quartile(Q1),median,third quartile(Q3),and "maximum).it can tell you about your outliers and what their values are.it also tell you if your data is symmetrical,how tightly your data is grouped,and ifand how your data is skewed.

```python
#Import necessary Libraries

import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv")
df
```

```python
df.describe()
```

```python
df.describe(include='O')
```

```python
#set the figure size of the graph 
```

```python
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=250
```

```python
#Selecting numerical columns
```

```python
num_cols=df.select_dtypes([int,float])

num_cols
```

```python
#matplotlib boxplot
```

```python
plt.boxplot(df['symboling'],notch=True, patch_artist=True)
```

```python
plt.boxplot(num_cols)
num_cols.head(2)
```

```python
num_cols.drop(columns=['price','engine-size']).boxplot()
```

```python
plt.boxplot(num_cols['city-mpg'])
plt.grid()
```
